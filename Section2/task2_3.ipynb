```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

```python
# Task 2.3 a)

xs = np.arange(0, 10000)
ys = [100*(int(p/1200 + 1)%2) for p in xs]

time4a, temp4a = np.loadtxt("thermal_4min_a.txt", delimiter='\t', skiprows=3).transpose()
time4b, temp4b = np.loadtxt("thermal_4min_b.txt", delimiter='\t', skiprows=3).transpose()

sin = lambda x, a, b, c, d: a*np.sin(b*x + c) + d
fit, cov = curve_fit(sin, time4a, temp4a, p0 = [20, 0.001, 1000, 50])
print(fit)

plt.plot(xs, ys)
plt.plot(time4a, temp4a)
#plt.plot(time4b, temp4b)
plt.plot(time4a, sin(time4a, *fit))
plt.show()
```
