```python
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
# Task 2.4 a)

for file in os.listdir():
    filename = os.fsdecode(file)
    if filename.endswith(".txt") and filename[13] != 'b':
        time, temp = np.loadtxt(filename, delimiter='\t', skiprows=3).transpose()
        plt.title(filename.lstrip("thermal_").rstrip(".txt"))
        plt.plot(time, temp)
        plt.xlabel("Time [0.1 s]")
        plt.ylabel("Temperature [deg C]")
        #plt.savefig(fname="output_" + filename.lstrip("thermal_").rstrip(".txt") + ".png", dpi=1000)
        plt.show()
```
