```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Task 1.2 b)
xs = np.arange(0, 500)
ys = [100*(int(p/120 + 1)%2) for p in xs]

plt.plot(xs, ys)
```

```python
# Task 1.2 c)
n_max = 100
b = np.append(np.array([0]), [100*(1-np.cos(np.pi*n))/(np.pi*n) for n in range(1, n_max+1)])

fs = lambda x: 50 + sum([b[n]*np.sin(2*np.pi*n*x/240) for n in range(1, n_max+1)])
y_fourier = fs(xs)

plt.plot(xs, ys)
plt.plot(xs, y_fourier)
```
