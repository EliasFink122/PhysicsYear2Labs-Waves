```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x1, y1 = np.loadtxt("Task1.3_Semicircle_low.txt", unpack=True, skiprows=1)
x2, y2 = np.loadtxt("Task1.3_Semicircle_high.txt", unpack=True, skiprows=1)
x3, y3 = np.loadtxt("Task1.3_Sine_low.txt", unpack=True, skiprows=1)
x4, y4 = np.loadtxt("Task1.3_Sine_high.txt", unpack=True, skiprows=1)

#plt.plot(x1, y1)
#plt.plot(x2, y2)
plt.plot(x3, y3)
#plt.plot(x4, y4)
```

```python
# Task 1.3 b)

# rectangular
area1_rect = sum([y1[i]*(x1[1]-x1[0]) for i in range(len(x1))])
area2_rect = sum([y2[i]*(x2[1]-x2[0]) for i in range(len(x2))])
print("Semi-circle rect: ", area1_rect, area2_rect)

area3_rect = sum([y3[i]*(x3[1]-x3[0]) for i in range(len(x3))])
area4_rect = sum([y4[i]*(x4[1]-x4[0]) for i in range(len(x4))])
print("Sine rect: ", area3_rect, area4_rect)

# trapezoid
area1_trap = (x1[-1] - x1[0])/(len(x1)) * (sum([y1[i] for i in range(len(x1))]) - 0.5*(y1[-1] + y1[0]))
area2_trap = (x2[-1] - x2[0])/(len(x2)) * (sum([y2[i] for i in range(len(x2))]) - 0.5*(y2[-1] + y2[0]))
print("Semi-circle trap: ", area1_trap, area2_trap)

area3_trap = (x3[-1] - x3[0])/(len(x3)) * (sum([y3[i] for i in range(len(x3))]) - 0.5*(y3[-1] + y3[0]))
area4_trap = (x4[-1] - x4[0])/(len(x4)) * (sum([y4[i] for i in range(len(x4))]) - 0.5*(y4[-1] + y4[0]))
print("Sine trap: ", area3_trap, area3_trap)
```
