```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Task 3.7 a)

data = np.loadtxt("waves_3_7_data.csv", delimiter=",", skiprows=1).transpose()

k = data[0]/40
omega = 2*np.pi*data[1]

plt.title("Dispersion relation")
plt.plot(k, omega, 'x')
plt.xlabel("$k\;[rad\,section^{-1}]$")
plt.ylabel("$\omega\;[krad\,s^{-1}]$")
plt.show()
```

```python
# Task 3.7 b)

v_group = np.array([(omega[i+1] - omega[i])/(k[i+1] - k[i]) for i in range(len(k) - 1)])

plt.title("Group velocity")
plt.plot(omega[:-1], v_group, 'x')
plt.xlabel("$\omega\;[krad\,s^{-1}]$")
plt.ylabel("$v_{group}\;[knodes\,s^{-1}]$")
plt.show()
```

```python
# Task 3.7 c)

v_phase = np.array([omega[i]/k[i] for i in range(len(k) - 1)])

plt.title("Phase and group velocity")
plt.plot(omega[:-1], v_phase, '^', label="Phase velocity")
plt.plot(omega[:-1], v_group, 'v', label="Group velocity")
plt.xlabel("$\omega\;[krad\,s^{-1}]$")
plt.ylabel("$v\;[knodes\,s^{-1}]$")
plt.legend()
plt.plot()
```

```python
# Task 3.7 d)

amp_in = data[2]
amp_out = data[3]
ratio = amp_out/amp_in

plt.title("Amplitude ratio")
plt.plot(omega, np.log(ratio), 'x')
plt.xlabel("$\omega\;[krad\,s^{-1}]$")
plt.ylabel("Amplitude ratio")
plt.show()
```
